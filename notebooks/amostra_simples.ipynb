```python
#importação das bibliotecas

import pandas as pd
import numpy as np
```

```python
#Adição da base de dados

base = pd.read_csv("base_de_dados/iris.csv", sep=",")
base.shape #Visualizando a quantidade de linhas e colunas
```

```python
base.head(10)
```

```python
np.random.seed(2743) #Definindo uma semente

valores = np.random.choice(a = [0, 1], size = len(base), replace = True, p = [0.7, 0.3])
```

A variável valores vai receber um array com um tamanho de 150 que é igual ao len(base), esse array será composto pelos números 0 e 1 que são as opções para serem escolhidas (a = [0, 1]),esses dois números poderão ser repitidos mais de uma vez pois o replace foi definido como True o que permite a reposição dos dados e por último definimos que o zero tera 0.7 (70%) de chance de ser selecionado e o um 0.3 (30%) de chance de ser selecionado.

```python
valores
```

```python
len(valores)
```

```python
amostra = base.loc[valores == 1]

amostra.shape
```

```python
#Outra forma para também gerar a amostra simples

tamanho_amostra = 40

novos_valores = np.random.choice(a = base.index, size = tamanho_amostra, replace = False)

amostra2 = base.loc[novos_valores]

amostra2
```
